### 인터랙티브 그래프 (interactive graph)
- 마우스 움직임에 반응하여 실시간으로 모양이 변하거나 데이터를 표시하는 그래프
- 인터랙티브 그래프를 만들면 그래프를 자유롭게 조작하면서 관심 있는 부분을 자세히 살펴볼 수 있음

### Plotly
- 인터렉티브한 시각화가 가능한 파이썬 그래픽 라이브러리
- 기본적인 시각화부터 통계, 재무, 지리 과학 및 3차원 그래프를 포함한 40개 이상의 차트 타입을 제공하는 오픈소스 라이브러리
- 기본적으로 주피터 노트북에 시각화가 가능하고 독립형 HTML 파일에 저장하여 웹 브라우저에서 확인하는 것이 가능
- Dash를 사용하여 순수 Python으로 구축된 웹 애플리케이션의 일부로 제공할 수 있는 대화형 웹 기반 시각화 가능
- https://plotly.com/graphing-libraries/

#### Plotly를 이용한 그래프 생성 방법
- graph_objects 모듈을 활용한 생성
  - 그래프를 하나하나 설명하고 직접 지정하며 제작하는 방법
- express 모듈을 활용한 생성
  - 그래프를 이미 제작되어있는 템플릿으로 빠르게 제작하는 방법
  - 매우 짧은 코드만으로 양질의 그래프를 제작할수있는 high-level 모듈로 graph_objects 기반으로 제작됨
  - graph_objects의 기능인 add_trace(), update_layout(), update_traces() 기능을 활용하여 세세한 튜닝이 가능

##### 각각의 장점과 단점
- graph_objects
  - 장점: 그래프를 세세하게 구성 가능
  - 단점: 코드가 길고 문법이 복잡하여 시간이 오래 걸리고 상대적으로 어려움
- express
  - 장점: 간단한 코드로 쉽게 그래프 표현 가능
  - 단점: 정해진 템플릿 외에 세세한 조정이 어려워 필요시 graph_objects 도움이 필요함

##### express 모듈로 사용 가능한 그래프 템플릿 (https://plotly.com/python/plotly-express/)
- Basics: scatter, line, area, bar, funnel, timeline
- Part-of-Whole: pie, sunburst, treemap, icicle, funnel_area
- 1D Distributions: histogram, box, violin, strip, ecdf
- 2D Distributions: density_heatmap, density_contour
- Matrix or Image Input: imshow
- 3-Dimensional: scatter_3d, line_3d
- Multidimensional: scatter_matrix, parallel_coordinates, parallel_categories
- Tile Maps: scatter_mapbox, line_mapbox, choropleth_mapbox, density_mapbox
- Outline Maps: scatter_geo, line_geo, choropleth
- Polar Charts: scatter_polar, line_polar, bar_polar
- Ternary Charts: scatter_ternary, line_ternary

#### 1. 산포 그래프(Scatter plot. 산점도) 만들기

##### 예제 데이터
- mpg.csv: 자동차 관련 정보를 담고 있는 데이터

```python
import pandas as pd

mpg = pd.read_csv('mpg.csv')
```

```python
import plotly.express as px # Plotly express 모듈 임포트

# 산포 그래프 함수 scatter()
px.scatter(data_frame = mpg, x = 'cty', y = 'hwy', color = 'drv')
# data_frame: 그래프를 그리는 데 사용할 DataFrame 객체
# x: x축에 표현할 컬럼(변수)
# y: y축에 표현할 컬럼(변수)
# color: 색상을 다르게 표현할 컬럼(변수)
```

##### 인터랙티브 기능 활용
- 산점도의 각 표식 (데이터 포인트)에 마우스 커서를 올리면 해당 데이터 값이 나타남
- 마우스를 드래그하면 x, y축의 범위가 바뀌고 그래프를 더블클릭하면 원래대로 되돌아감
- 범례 항목을 클릭하면 특정 범주의 표식을 제외하거나 포함할 수 있음
- 그래프 우상단의 버튼을 활용하여 줌인, 줌아웃, 패닝 등의 기능 활용 가능

##### 그래프 크기 조절하기

```python
# width, height 매개변수 값을 설정하여 그래프의 가로, 세로 크기 조절
px.scatter(data_frame = mpg, x = 'cty', y = 'hwy', color = 'drv', width=600, height=500)
```

##### 이 외 다양한 기능이 있음

##### 그래프 제목

```python
import plotly.express as px

# 데이터 불러오기
# plotly.express.data 패키지: plotly에서 시연, 교육, 테스트 목적으로 사용하는 내장 데이터셋
# tips(): 레스토랑 tip 데이터
df = px.data.tips()

#그래프 그리기
fig = px.scatter(df, x="total_bill", y="tip", title="Total bill and tip scatter plot")
# title 매개변수: 그래프 제목

fig.show()
```

```python
# 타이틀 위치 설정
fig.update_layout(
                 title_x = 0.5,
                 title_y = 0.9,
                 title_xanchor = "center",
                 title_yanchor = "middle")

# title_x = 가로 축의 좌표로 0은 맨 왼쪽 1은 맨 오른쪽
# title_y = 세로 축의 좌표로 0은 맨 아래 1은 맨 윗쪽
# title_xanchor = 좌표를 중심으로 타이틀을 왼쪽, 또는 가운데, 오른쪽에 놓을지 설정
# title_yanchor = 좌표를 중심으로 타이틀을 위에, 또는 가운데, 아래에 놓을지 설정

fig.show()
```

##### 축 제목

```python
fig = px.scatter(df, x = "total_bill", y = "tip",
                    labels = dict(total_bill = "Total Bill ($)", tip = "Tip ($)"))
# labels 매개변수: x축, y축 제목(레이블) 값을 dict 객체로 할당

fig.show()
```

##### 축 범위 

```python
fig = px.scatter(df, x = "total_bill", y = "tip",
                    labels = dict(total_bill = "Total Bill ($)", tip = "Tip ($)"))

# 축 범위 지정
# update_xaxes(): x축 범위를 업데이트 할 때 사용
# update_yaxes(): y축 범위를 업데이트 할 때 사용
# range 매개변수: 범위의 최소값과 최대값을 리스트로 할당
fig.update_xaxes(range=[0, 60])
fig.update_yaxes(range=[0, 15])

fig.show()
```

##### 눈금(tick) 설정

```python
fig = px.scatter(df, x = "total_bill", y = "tip",
                    labels = dict(total_bill = "Total Bill ($)", tip = "Tip ($)"))

# 눈금(tick) 생성
# update_xaxes(): x축 tick을 업데이트 할 때 사용
# update_yaxes(): y축 tick을 업데이트 할 때 사용
# ticks 매개변수: "inside" (눈금을 그래프 안쪽에 생성) 또는 "outside" (눈금을 그래프 바깥쪽에 생성) 
fig.update_xaxes(ticks="outside")
fig.update_yaxes(ticks="inside")

fig.show()
```

```python
# 눈금 간격 변경

fig.update_xaxes(dtick=5)
fig.update_yaxes(dtick=1)
# dtick 매개변수: 눈금 간격으로 설정할 값을 할당

fig.show()
```

##### 그리드 설정
- 기본 그리드: 큰 단위. 자동으로 보여짐
- minor 그리드: 작은 단위로 촘촘하게 나뉨. 따로 코드를 추가해야 표시됨

```python
# 기본 그리드, minor 그리드 시각화
fig.update_xaxes(showgrid=True, minor_showgrid=True)
fig.update_yaxes(showgrid=True, minor_showgrid=True)
# showgrid 매개변수: 그리드 시각화 여부 결정
# minor_showgrid 매개변수: minor 그리드 시각화 여부 결정

fig.show()
```

##### 범례 지정

```python
import plotly.express as px

# 데이터 불러오기
df = px.data.tips()

# 그래프 그리기
fig = px.scatter(df, x="total_bill", y="tip", color="time")
# color 매개변수: 색상을 다르게 구분할 컬럼(변수)명을 할당하면 이에 대해 자동으로 범례(legend)가 추가됨

fig.show()
```

```python
#범례 삭제하기

# express 모듈을 사용하는 경우 범례가 자동으로 추가되므로 삭제할 경우 update_layout() 함수를 사용하여 showlegend 매개변수 값을 False로 할당
fig.update_layout(showlegend=False)

fig.show()
```

```python
fig.update_layout(showlegend=True)

fig.show()
```

```python
# 범례 위치 지정하기

fig.update_layout(
    legend_x=0.01,
    legend_y=0.99,
    legend_xanchor="left",
    legend_yanchor="top"
)

fig.show()
```

##### 마커(점) 크기 다르게 하기

```python
# 그래프 그리기
fig = px.scatter(df, x="total_bill", y="tip", color="time", size="size")
# size 매개변수: 마커 크기를 조정할 컬럼 이름

fig.show()
```

##### 2. 막대 그래프 (bar plot) 만들기

```python
# mpg 데이터에서 자동차 종류별 빈도 구하기

mpg = pd.read_csv('mpg.csv')

df = mpg.groupby('category', as_index = False).agg(n=('category', 'count'))
print(df)
```

```python
# 막대 그래프 그리기

# 막대 그래프 함수 bar()
fig = px.bar(data_frame = df, x = 'category', y = 'n', color = 'category')

fig.show()
```

##### 3. 선 그래프 (line plot) 만들기

##### 예제 데이터

```python
import plotly.express as px

# 데이터 불러오기
# plotly.express.data 패키지: plotly에서 시연, 교육, 테스트 목적으로 사용하는 내장 데이터셋
# gapminder(): 국가별 년도별 life expectancy, GDP per capita, 그리고 population에 대한 데이터
# GDP per capita (1인당 GDP): 한 국가의 총 국내총생산(GDP)을 전체 인구로 나눈 값
# ISO alpha: ISO 3166-1 표준에 정의된 국가 코드

df = px.data.gapminder()

print(df)
```

```python
# gapminder 데이터에서 country가 'Canada'인 행만 추출
data = df.query("country=='Canada'")

# 선 그래프 그리기

# 선 그래프 함수 line()
fig = px.line(data_frame= data, x="year", y="lifeExp")

fig.show()
```

##### 라인 색 다르게 표시하기

```python
# 사용할 데이터 
data = df.query("continent=='Oceania'")

fig = px.line(data, x="year", y="lifeExp", color='country')

fig.show()
```

##### 마커를 같이 표시하기

```python
fig = px.line(data, x="year", y="lifeExp", color='country', markers=True)

fig.show()
```

##### 마커 스타일 다르게 표시하기

```python
fig = px.line(data, x="year", y="lifeExp", color='country', markers=True, symbol="country")

fig.show()
```

##### 여러 개로 나눠 그리기 (Facet)

```python
fig = px.line(data, x="year", y="lifeExp", facet_col="country")
# facet_col: 컬럼을 구분하여 나눠 그리기
# facet_row: row(행)을 구분하여 나눠 그리기

fig.show()
```
